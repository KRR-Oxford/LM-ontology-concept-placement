Training Bi-encoder and Cross-encoder

This shell script executes:
(i) Edge search with Edge-Bi-encoder training, evaluation
(ii) Edge enrichment
(iii) Edge selection: Edge-Cross-encoder

The script is Edge-Bi-enc+Cross-enc.sh \$1 \$2 \$3 \$4 \$5 \$6 \$7 \$8 \$9

Parameters (marked with \$1 to \$9):
1. subset string: "Disease" or "CPP"
2. whether to use context
3. top-k value
4. number of edge seeds before edge enrichment into top-k
5. biencoder training batch size
6. whether to use debugging mode in training (training biencoder with first debug_max_lines, which have default value as 1000)
7. whether to use a sample of first n rows of mention-edge pairs for creating data for cross-encoder.
8. n rows mention-edge pairs for creating data for cross-encoder
9. whether to use debugging mode in inference (for in-KB, inference with only a random 100 data lines; for out-of-KB, still inference will all data lines).

````python
# using debugging mode and using first 1000 mention-edge pairs for creating data for cross-encoder - this can take around 1 hour with the T4 GPU.
# setting debugging variables
debugging_mode_training = True
# construct shell command with the debugging variables
debug_tr_str = str(debugging_mode_training).lower()
!bash Edge-Bi-enc+Cross-enc.sh Disease true 50 25 16 {debug_tr_str} true 1000 #> results_log_Disease_edge_biencoder_top50_1k.txt
#!bash Edge-Bi-enc+prompt-generation.sh Disease true 10 5 16 > results_log_Disease_edge_biencoder_top10_bs16_final.txt
````

Display an intermediary prompt below (the first prompt in the test-NIL dataset).

Note that the intermediary prompt is not the final promot - which is later preprocessed with

```
from blink.prompting.Llama_2_finetuning import prompt_concat_ctx_and_men, prompt_id_removal, prompt_correction
from blink.prompting.Llama_2_prompting import format_prompt
prompt = prompt_correction(prompt) # correct prompt
prompt = prompt_concat_ctx_and_men(prompt) # concatenate context and mention in the prompt (i.e. use the non-separated paragraph with mention marked as *mention*.)
prompt = prompt_id_removal(prompt) # remove ids (and parentheses) in prompt
```

For Llama 2 models, this further processing is applied
```
prompt_formatted = format_prompt(prompt,cot=True) # format prompt for Llama 2 (same as in fine-tuning), here cot means using explaination before getting the answer
```

For the final version of a prompt, see Section 4.3 on page 8-9 in the [paper](https://arxiv.org/pdf/2402.17897).
